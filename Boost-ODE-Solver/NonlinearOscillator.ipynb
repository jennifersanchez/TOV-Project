### then the scipy ode solver

```python
# scipy ODE solver 
# help found here: (https://sam-dolan.staff.shef.ac.uk/mas212/notebooks/ODE_Example.html)
from scipy.integrate import odeint
def dU_dx(U,x):
    return [U[1], -0.4*U[1]-4*U[0]]

U0 = [1,0] # your initial conditions
xs = np.linspace(0, 15, 151) # first two values are your y-lim and last is your K value (how smooth the curve will be)
Us = odeint(dU_dx, U0, xs) # intergrating?
ys = Us[:,0]

plt.xlabel("Time (s)")
plt.ylabel("x(m)")
plt.title("Damped harmonic oscillator")
plt.plot(xs,ys, label = 'scipy solution')
plt.grid()
plt.legend()
```
